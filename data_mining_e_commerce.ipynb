#En iyi satış yapılan ay hangisidir? Ne kadar kazanılmıştır?



aylar için yeni sütun ekliyoruz

```python
df['Month'] = pd.DatetimeIndex(df['order_date']).month
```

yeni df

```python
df
```

```python
df.dtypes
```

```python
dff = df[df['item_total'] != "item_total"]


dff.dtypes
```

```python
dff['item_total'] = dff['item_total'].str.replace('₹', '')
dff['item_total'] = dff['item_total'].str.replace(',', '.')
dff['item_total'] = dff['item_total'].str.replace('.00', '')


dff['item_total'] = dff['item_total'].astype(float)
```

```python
dff.dtypes
```

```python
dff
```

```python
dff["Sales"]=dff["quantity"]*dff["item_total"]
dff
```

```python
df_month=dff.groupby(["Month"])['Sales'].sum().sort_values(ascending=False)
df_month=df_month.to_frame()
df_month
```

```python
fig = plt.figure(figsize = (15, 8))
Month=["Dec","Oct","Nov","Jan","Sep","Feb","Aug"] 
df_month["Month"]=Month
sns.barplot(x = "Month",y = "Sales",data = df_month)
plt.title("En iyi satış yapılan ay")
plt.show()
```

nek adar kazanmış

```python
Dec = dff[df.Month == 12]
sales = Dec.Sales
print("Total Sales in Dec= ",sales.sum())
```
